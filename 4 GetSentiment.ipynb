## Use regular expressions to remove trailing text in lyrics

```python
import re
def remove_tail(text):
    rem = 'You might also likeEmbed'
    ind = -len(rem)
    if text[ind:] == rem:
        return(text[:ind])
    else:
        reg = 'You might also like.Embed'
        ind_reg = -len(reg)
        x=re.search(reg,text[ind_reg:])
        if x == None:
            return(text)
        else:
            return(text[:ind_reg])
```

```python
df['lyrics'] = df.lyrics.apply(remove_tail)
# df['lyrics'] = df.lyrics.apply(first_step_normalizer)
```

## Use TextBlob to return sentiment of each set of lyrics

```python
from textblob import TextBlob
def get_sentiment(text):
    return(TextBlob(text).sentiment.polarity)
```

```python
df['sentiment'] = df.lyrics.apply(get_sentiment)
```
